# M 1.2 Distribution analysis

### Method of Moments

- I implemented a normal distribution to calculate the method of moments
- As the mean values of normalized death count for counties is a floating point I consider normal distribution
- For the calculation of Method of Moments in the normal distribution we consider the 'mu' value as the mean and 'sigma' as the standard deviation of the considered data.
- Range of the plot is -3 sigma to +3 sigma and so distribution covers 99% of the data.

**Method of Moments for the Normal distribtion is always the Mean value of the distribution**

```python
mu = nc_mean
sigma = nc_std
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
```

```python
mu1 = ky_mean
sigma1 = ky_std
x1 = np.linspace(mu1 - 3*sigma1, mu1 + 3*sigma1, 100)
```

```python

fig = go.Figure()
fig.add_trace(go.Histogram(name = 'KY', x=df_ky['Norm_Deaths'], nbinsx = 40, marker = {'color' : 'orange'}, histnorm = 'probability'))
fig.add_trace(go.Histogram(name = 'NC', x=df_nc['Norm_Deaths'], nbinsx = 30, marker = {'color' : 'blue'}, histnorm = 'probability'))


fig.add_trace(go.Scatter({'x': x, 'y' : stats.norm.pdf(x, mu, sigma), 'name' : 'NC MoM PDF'}, marker = {'color' : 'blue'}))
fig.add_trace(go.Scatter({'x' : x1, 'y' : stats.norm.pdf(x1, mu1, sigma1), 'name' : 'KY MoM PDF'}, marker = {'color' : 'orange'}))

#traces.append({'x': X1, 'y': Y1, 'name': 'D1'})

fig.update_xaxes(title_text="Number of Opioid Deaths for 100k population")
fig.update_yaxes(title_text="Number of Counties")
fig.update_layout(title_text = "Distribution of per capita deaths for different counties in NC and KY")
fig.update_layout(barmode='overlay')
fig.update_traces(opacity = 0.75)
fig.add_vline(x= ky_mean, line_color='red', line_dash="dash", annotation_text=f'KY MEAN = {ky_mean}', annotation_position='top right')
fig.add_vline(x= nc_mean,line_color='blue', line_dash="dash", annotation_text=f'NC MEAN = {nc_mean}', annotation_position='top left')
#fig.show()




fig.write_image("../../../../images/vijay/NC&KY_hist_dist_mom.png")
Image(filename='../../../../images/vijay/NC&KY_hist_dist_mom.png')
```

### Two-Sample T-Test

- **A two-sample t-test investigates whether the means of two independent data samples differ from one another.**
- **null hypothesis is that, The mean values of both the considered groups (Divided each observation into low and high) are same**
- **Alternate hypothesis can be considered as the mean values for both the groups are not simimlar**




- **Comparing the distributions for the hypothesis test**
- **Only plotted the distribution for variable I (Opioid Dispensing Rate)**
- **Divided the variables data into two (High and Low) and plotted the Histograms for the data along with their respective means of the data.**

```python

fig = go.Figure()
fig.add_trace(go.Histogram(name = 'low_Opiod_Dispensing_Rate', x=low['Opiod_Dispensing_Rate'], nbinsx = 50, marker = {'color' : 'orange'}))
fig.add_trace(go.Histogram(name = 'high_Opiod_Dispensing_Rate', x=high['Opiod_Dispensing_Rate'], nbinsx = 50, marker = {'color' : 'blue'}))
#fig.add_vline(x= mean_us,line_color='red', annotation_text='mean', annotation_position='bottom')



fig.update_xaxes(title_text="Opiod_Dispensing_Rate")
fig.update_yaxes(title_text="Counts")
fig.update_layout(title_text = "histogram for low and high low_Opiod_Dispensing_Rate")
fig.update_layout(barmode='overlay')
fig.update_traces(opacity = 0.75)
fig.add_vline(x= low_mean, line_color='red', line_dash="dash", annotation_text=f'LOW MEAN = {low_mean}', annotation_position='top left')
fig.add_vline(x= high_mean,line_color='blue', line_dash="dash", annotation_text=f'HIGH MEAN = {high_mean}', annotation_position='top right')
#fig.show()


fig.write_image("../../../../images/vijay/histogram for low and high low_Opiod_Dispensing_Rate.png")
Image(filename='../../../../images/vijay/histogram for low and high low_Opiod_Dispensing_Rate.png')
```

- Two samples are extracted by dividing the observations into two categories high and low
- Hypothesis test is done using two-sample T- test(As we are considering 2 samples for each observation/ variable)
- **We can conduct a two-sample t-test by passing with the stats.ttest_ind() function**

```python
stats.ttest_ind(a= low.Norm_Deaths,
                b= high.Norm_Deaths,
                equal_var=False)
```

- for each variable we add a label and divide the dataframe accordingly. For the divided dataframe we compare the samples with the opioid mortality rates.
- sample 'a' will be for the low values and sample '1' will be for high values
- same process is done for all the five variables

- **for variable I, obtained 'p' value is almost equal to zero, so we reject the null hypothesis**
- if 'p' value greater than 0.05 then null hypothesis is accepted and we can say their is a relation between the mean values of the both the 
- The mean values of both the considered groups are almost similar when the 'p' value greater than 0.05 then the test will fail to reject.

- variable II
- Life expectancy raw value

```python
mean1 = df['Life expectancy raw value'].mean()
max1 = df['Life expectancy raw value'].max()
min1 = df['Life expectancy raw value'].min()
mean1, max1, min1
```

```python
df['Life_expectancy_raw_value_label']= pd.cut(df['Life expectancy raw value'], bins = [min1, mean1, max1], labels = [0, 1])
low1 = df[df['Life_expectancy_raw_value_label'] == 0]
high1 = df[df['Life_expectancy_raw_value_label'] == 1]
stats.ttest_ind(a= low1.Norm_Deaths,
                b= high1.Norm_Deaths,
                equal_var=False)

```

- 'p' value is almost zero so we reject the null hypothesis for variable II

- variable III

```python
mean2 = df['Drug overdose deaths raw value'].mean()
max2 = df['Drug overdose deaths raw value'].max()
min2 = df['Drug overdose deaths raw value'].min()
mean2, max2, min2
```

```python
df['Drug_overdose_deaths_raw_value_label']= pd.cut(df['Drug overdose deaths raw value'], bins = [min2, mean2, max2], labels = [0, 1])
low2 = df[df['Drug_overdose_deaths_raw_value_label'] == 0]
high2 = df[df['Drug_overdose_deaths_raw_value_label'] == 1]
stats.ttest_ind(a= low2.Norm_Deaths,
                b= high2.Norm_Deaths,
                equal_var=False)

```

- 'p' value is greater than 0.05 so we fail to reject the null hypothesis for variable III
