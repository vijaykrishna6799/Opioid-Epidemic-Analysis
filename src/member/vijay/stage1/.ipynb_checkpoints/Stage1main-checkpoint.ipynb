# M3.1 Merging

- Merging Dataframes using pandas merge
- https://stackoverflow.com/questions/20375561/joining-pandas-dataframes-by-column-names

```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv("../../../../data/2019-Opioid_Rate.csv",encoding='latin-1')
# data frame df1 is opioid dispensing rates
df1
```

### Removing unwanted columns

```python
df1=df1.drop(['County'], axis=1)
df1
```

```python
df2=pd.read_csv("../../../../data/underlying_cause_of_death.csv")

df2 = df2[['County', 'County Code', 'Deaths', 'Population']] # using only the required columns

df2['Norm_Deaths']= round((df2['Deaths']/df2['Population'])*100000) # creates a new column with name 'Norm_Deaths' for 100,000 population


df2 = df2.groupby(['County', 'County Code', 'Population'], as_index=False)[["Deaths","Norm_Deaths"]].sum() # grouping by to extract county wise data to extraxt normalized death rates
```

```python
df2
# using same method on the dataframe as in the team task to take the normalized deaths per 100,000 population
```

```python
df12=pd.merge(df1, df2, left_on='FIPS', right_on='County Code')
df12
#Merging both dataframe (As we have different column names we use left_on and right_on)
```

```python
df3=pd.read_csv("../../../../data/County_Health_Ranking.csv")
df3
```

```python
df123=pd.merge(df12,df3,left_on='FIPS', right_on='5-digit FIPS Code')
df123

#merging output of dataframe 1,2 with 3
```

```python
df123.sort_values(by=['Norm_Deaths'],ascending=False).head(10)
```

```python
df123.to_csv('123Merged.csv',index=False)
```
